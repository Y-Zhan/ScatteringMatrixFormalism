```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm
```

```python
# operators
sigma = np.matrix([[0, 0], [1, 0]])
numPts = 1000
numdt = 1000
Tp = 1.0
dt = Tp/numdt
gamma = 1.0
Omega_0 = np.linspace(0, 10, numPts)
```

```python
ground = np.matrix([[0], [1]])
P0gResult = []
for ii in range(numPts):
    H_eff = -1.0j*gamma/2*np.dot(np.conjugate(sigma),sigma) + Omega_0[ii]*(sigma + np.conjugate(sigma))
    U_eff = np.matrix([[1.0, 1.0], [1.0, 1.0]])
    for ndt in range(numdt):
        U_effTemp = expm(-1.0j*dt*H_eff)
        U_eff = np.dot(U_effTemp, U_eff)
    P0g = np.dot(ground.T, np.dot(U_eff, ground))
    P0gResult.append(P0g)
```
