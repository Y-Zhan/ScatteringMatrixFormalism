```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
```

```python
num_pts = 801
t = np.linspace(0, 4, num_pts)
y = 1*np.sin(2*np.pi*25*t) + 2*np.sin(2*np.pi*50*t)
y_fft = np.fft.fftshift(np.fft.fft(y))/(num_pts - 1)
f_s = (num_pts - 1)/4
freq = np.linspace(0, f_s, num_pts) - f_s/2
```

```python
plt.figure()
plt.plot(freq, np.abs(y_fft))
```
